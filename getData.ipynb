```python
from shapely.geometry import Polygon
from shapely.geometry import Point
from shapely.geometry import box
import pandas as pd
import geopandas as gp
import numpy as np

def getpoint(coordinates):
    x=coordinates[0]
    y=coordinates[1]
    return Point(x,y)

def BOX(polygon):
    poly=Polygon(polygon)
    b=poly.bounds
    BOX=box(b[0],b[1],b[2],b[3])
    return BOX

def towards(shp):
    DataSet = list()
    polygons=shp['geometry']
    ward_names=shp['NAME']
    b_names=shp['BOROUGH']
    for i in range(len(shp)):
        DataSet.append(dict())
        DataSet[i]['Ward'] = ', '.join([ward_names[i],b_names[i]])
        DataSet[i]['Polygon'] = Polygon(np.array(polygons[i].exterior))
        DataSet[i]['Boundary'] = BOX(np.array(polygons[i].exterior))
    return DataSet

#Pass the tweets list to the above function to create a DataFrame

shp = gp.GeoDataFrame.from_file('Ward_London.shp')
wards=towards(shp)
print(wards)
```
